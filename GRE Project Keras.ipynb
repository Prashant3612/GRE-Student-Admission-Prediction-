```python
x=df.drop(columns=['Chance of Admit '])
y=df['Chance of Admit ']
```

# Model Building using Tensorflow

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.2,random_state=42)
```

```python
from sklearn.preprocessing import MinMaxScaler
mms=MinMaxScaler()
x_train_scaled=mms.fit_transform(x_train)
x_test_scaled=mms.transform(x_test)
```

```python
x_test
```

```python
x_train
```

```python
from tensorflow import keras
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense
from sklearn.metrics import accuracy_score, r2_score
from tensorflow.keras.layers import Dropout
```

```python
model=Sequential()
model.add(Dense(50,activation='relu',input_dim=7))
model.add(Dropout(0.2))
model.add(Dense(1,activation='linear'))
```

```python
model.summary()
```

```python
model.compile(loss='huber_loss',optimizer='adam',metrics=['mean_absolute_error'])
history=model.fit(x_train_scaled,y_train,epochs=100,validation_split=0.2, verbose=0)
```

```python
y_pred=model.predict(x_test_scaled)
r2_score(y_test,y_pred)
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
```
